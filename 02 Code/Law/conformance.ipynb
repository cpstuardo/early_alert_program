```python
sit_sem_susp = [
    'SUSPESTUDIOS - Suspensión de estudios',
    'SUSPESTUDIOS - Suspensión de estudios (Reintegro)',
    'SUSPSALUD - Suspensión por salud',
    'SUSPSALUD - Suspensión por salud (Reintegro)',
    'SUSPSANCACADEM - Suspensión sanción académ.',
    'SUSPFUERZMAYOR - Suspensión por fuerza mayor',
    'ANULESTUDIOS - Anulación de estudios'
]
```

### Vía de admisión

```python
costos_PSU = []
costos_BEA = []
costos_TI = []
costos_PACE = []
costos_otros = []

fitness_PSU = []
fitness_BEA = []
fitness_TI = []
fitness_PACE = []
fitness_otros = []

for estudiante in estudiantes_d:
    if estudiante in novig_info_d:
        info = novig_info_d
    else:
        info = vig_info_d
    
    if info[estudiante]['PERIODO ADMISION'] == 201920:
        if info[estudiante]['TIPO ADMISION'] == '60 - PSU':
            costos = costos_PSU
            fitness = fitness_PSU
        elif info[estudiante]['TIPO ADMISION'] == '18 - Beca Excelencia Académica':
            costos = costos_BEA
            fitness = fitness_BEA
        elif info[estudiante]['TIPO ADMISION'] == '12 - Talentos e Inclusión':
            costos = costos_TI
            fitness = fitness_TI
        elif info[estudiante]['TIPO ADMISION'] == '56 - PACE':
            costos = costos_PACE
            fitness = fitness_PACE
        else:
            costos = costos_otros
            fitness = fitness_otros
        
        costo = 0
        for i in range(1,6):
            if i in estudiantes_d[estudiante].keys():
                if estudiantes_d[estudiante][i]['SITUACION ACADEMICA'] in sit_sem_susp:
                    costo += 0.5
                else:
                    if 'ALERTA' in estudiantes_d[estudiante][i].keys():
                        costo += 1
        costos.append(costo)
        fitness.append(1-costo/len(estudiantes_d[estudiante]))
        

# PSU
meanc_PSU = statistics.mean(costos_PSU)
stdevc_PSU = statistics.pstdev(costos_PSU)
min_PSU = min(costos_PSU)
max_PSU = max(costos_PSU)
meanf_PSU = statistics.mean(fitness_PSU)
stdevf_PSU = statistics.pstdev(fitness_PSU)

# BEA
meanc_BEA = statistics.mean(costos_BEA)
stdevc_BEA = statistics.pstdev(costos_BEA)
min_BEA = min(costos_BEA)
max_BEA = max(costos_BEA)
meanf_BEA = statistics.mean(fitness_BEA)
stdevf_BEA = statistics.pstdev(fitness_BEA)


# TI
meanc_TI = statistics.mean(costos_TI)
stdevc_TI = statistics.pstdev(costos_TI)
min_TI = min(costos_TI)
max_TI = max(costos_TI)
meanf_TI = statistics.mean(fitness_TI)
stdevf_TI = statistics.pstdev(fitness_TI)
        
# PACE
meanc_PACE = statistics.mean(costos_PACE)
stdevc_PACE = statistics.pstdev(costos_PACE)
min_PACE = min(costos_PACE)
max_PACE = max(costos_PACE)
meanf_PACE = statistics.mean(fitness_PACE)
stdevf_PACE = statistics.pstdev(fitness_PACE)

# Otros
meanc_otros = statistics.mean(costos_otros)
stdevc_otros = statistics.pstdev(costos_otros)
min_otros = min(costos_otros)
max_otros = max(costos_otros)
meanf_otros = statistics.mean(fitness_otros)
stdevf_otros = statistics.pstdev(fitness_otros)
```

```python
col_names = ["Tipo admisión", "Núm. estudiantes","Costo promedio", "Costo mínimo", "Costo máximo", "Fitness promedio", "Desv. fitness"]
data = [["PSU", len(costos_PSU), meanc_PSU, min_PSU, max_PSU, meanf_PSU, stdevf_PSU],
        ["BEA", len(costos_BEA), meanc_BEA, min_BEA, max_BEA, meanf_BEA, stdevf_BEA],
        ["TI", len(costos_TI), meanc_TI, min_TI, max_TI, meanf_TI, stdevf_TI],
        ["PACE", len(costos_PACE), meanc_PACE, min_PACE, max_PACE, meanf_PACE, stdevf_PACE],
        ["Otros", len(costos_otros), meanc_otros, min_otros, max_otros, meanf_otros, stdevf_otros]
       ]
  
print(tabulate(data, headers=col_names))
```
